```
import time
import datetime
import json,csv

headers = True
filein = "/content/gdrive/My Drive/review.json"
fileout = "/content/gdrive/My Drive/review_final.csv"

with open(filein,encoding="UTF-8") as jsonf, open(fileout,"w") as csvf:
    for line in jsonf:
        data = json.loads(line)
        # 필요없는 key값삭제하면서 나중에 읽어올때, 시간 절약
        data.pop("review_id")
        data.pop("date")
        data.pop("text")
        data.pop("useful")
        data.pop("funny")
        data.pop("cool")
        
        
        if headers:
            keys = []
            for k,v in data.items():
                # key 값은 column 으로 value 는 값으로 저장
                keys.append(k)
            writer = csv.DictWriter(csvf,fieldnames=keys)
            writer.writeheader()
            headers = False
        writer.writerow(data)

print("success")
```
